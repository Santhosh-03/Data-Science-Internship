## 1. The Captain's Room



```python
k = int(input()) # getting the input
arr = list(map(int, input().split())) # getting a list of elements seperated by space
count_dict = {} # empty dictionary to store the count of each element in the list

for num in arr:
    if num not in count_dict: # new number is taken from the list
        count_dict[num] = 1 # initialising the value as 1 for new key
    else:
        count_dict[num] += 1 # incrementind the value for key num
        
for num in count_dict:
    if count_dict[num] == 1: # checking the value for a key which the required number
        print(num)
        break
```

## 3. Find the Runner-Up Score!



```python
n = int(input()) # getting the number of elements for the list
arr = list(map(int, input().split()))[:n] # getting the list of n elements seperated by a space

arr.sort(reverse = True) # sorting the list in decending order
max_score = max(arr) # getting the maximum value of the list i.e the first element
for num in arr: # iterating each element in the list
    if num < max_score: # True when the num is the second largest in the list
        print(num)
        break # coming out of the loop after finding the second largerst element
```

## 4. Nested Lists



```python
n = int(input()) # getting the number of students from the user
students_list = [] # empty list
for _ in range(n): # getting n inputs
    name = input() 
    mark = float(input())
    students_list.append([name, mark]) # updating the student_list with the info given as input

# getting the second lowest mark by removing duplicated using set and sorting it
second_mark = sorted(list(set([mark for name, mark in students_list])))[1]
for name, mark in sorted(students_list):
    if mark == second_mark: # if given mark is equal to second lowest mark
        print(name) # printing the name
```

## 5. Finding the percentage



```python
n = int(input()) # getting the number of inputs
student_dict = {} # empty dictionary to store the inputs

for _ in range(n):
    details = list(map(str, input().split())) # getting the list of name and marks seperated by space
    student_dict[details[0]] = list(map(float, details[1:]))
    # first element in the list is name         from the first index the elements are marks 
    # so it is the key                          which are values
    
query_name = input() # getting the required name to find the average
for name in student_dict.keys(): # iterating through the list of key names in the dictionary
    if name == query_name: # if cuurent name is equal to the queried name
        print('{0:.2f}'.format(sum(student_dict[name])/len(student_dict[name]))) # printing the average of marks for that name
```

## 6. Lists



```python
n = int(input()) # getting the range of operation to perform
ls = []

for _ in range(n):
    operation = list(map(str, input().split())) # getting the required operation and its inputs
    if operation[0] == 'insert':
        pos, val = int(operation[1]), int(operation[2]) # changing the type of position and value to int
        ls.insert(pos, val)# inserting the val at pos
    elif operation[0] == 'append':
        val = int(operation[1]) # changing the type value to int
        ls.append(val) # appending the element at the end of the list
    elif operation[0] == 'remove':
        val = int(operation[1]) # changing the type value to int
        ls.remove(val) # removing the mentioned value from the list
    elif operation[0] == 'sort':
        ls.sort() # sorting the list in ascending order
    elif operation[0] == 'reverse':
        ls.reverse() # reversing the content of the list
    elif operation[0] == 'pop':
        ls.pop() # removing the last element of the list
    elif operation[0] == 'print':
        print(ls) # printing the contents of the list
```

## 8. Introduction to Sets



```python
def average(array):
    unique_array = list(set(array)) # getting the unique elements of the list using set()
    average = round(sum(unique_array)/len(unique_array),3) # finding the average and rounding it to 3 decimal places
    return average # returning the obtained average
if __name__ == '__main__':
    n = int(input()) # getting the size of the list
    arr = list(map(int, input().split())) # getting the elements of the list
    result = average(arr) # calling the function with the list arr as parameter
    print(result) # printinf the returned value from the function
```

## 10. Symmetric Difference



```python
n = int(input()) # size of set_A
set_A = set(map(int, input().split())) # getting the elements of set_A
m = int(input())# size of set_B
set_B = set(map(int, input().split())) # getting the elements of set_A

p = set_B.difference(set_A) # values present in set_B but not in set_A
q = set_A.difference(set_B) # values present in set_A but not in set_B
r = p.union(q) # combining both elements of p and q into a single unit
 
for num in sorted(r): # iterating through the sorted list
    print(num) # printing each number in r
```

## 12. Set .discard(), .remove() & .pop()



```python
size = int(input()) # size of the initial set 'set_'
set_ = set(map(int, input().split())) # getting the elements for the set
no_of_ops = int(input()) # number of operations to be done on the set 'set_'

for _ in range(no_of_ops):
    op = list(map(str, input().split())) # getting the operation and its required parameters
    if op[0] == 'remove': # checking whether operation in remove 
        val = int(op[1]) # converting the type of the element remove to integer
        set_.remove(val) # removing the value
    elif op[0] == 'discard': # checking whether operation in discard
        val = int(op[1]) # converting the type of the element discard to integer
        set_.discard(val) # discarding the value
    elif op[0] == 'pop': # checking whether operation in pop
        set_.pop() # removes the first element from the set
        
print(sum(list(set_)))
```

## 17. Set Mutations



```python
size = int(input()) # getting the size of elements for set_A
set_A = set(map(int, input().split())) # getting the values for set_A

no_of_ops = int(input()) # number of operations to be performed

for _ in range(no_of_ops):
    op, n = map(str, input().split()) # getting the operation and range of the second set
    if op == 'intersection_update':
        set_B = set(map(int, input().split())) # getting the values for set_B
        # getting the elements present in both the sets and 
        # storing it in the initial set set_A
        set_A.intersection_update(set_B) 
        
    elif op == 'update':
        set_B = set(map(int, input().split())) # getting the values for set_B
        # updating the initial set set_A by adding the elements from set_B into set_A
        set_A.update(set_B)
        
    elif op == 'symmetric_difference_update':
        set_B = set(map(int, input().split())) # getting the values for set_B
        # updating the initial set set_A by taking the elements that are not present in both the sets
        set_A.symmetric_difference_update(set_B)
        
    elif op == 'difference_update':
        set_B = set(map(int, input().split())) # getting the values for set_B
        # updates the initial set by removing the elements of set_B that are present in set_A
        set_A.difference_update(set_B)
        
print(sum(list(set_A)))
```
