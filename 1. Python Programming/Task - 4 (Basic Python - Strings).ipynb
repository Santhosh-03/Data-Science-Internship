## 1. sWAP cASE



```python
def swap_case(given_string): # fuction defenition
    new_string = '' # empty string
    for char in given_string: # for every character in the given string 
        if char.isupper(): # checking whether it is in upper case
            new_string += char.lower() # appending the lower case of that character
        elif char.islower(): # checking whether it is in lower case
            new_string += char.upper() # appending the upper case of that character
        else: # other characters
            new_string += char # just appending to the new string
    return new_string
    
    
if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

## 2. String Split and Join



```python
def split_and_join(line): # function defenition
    line_list = line.split() # splitting the input of the function at each space
    new_line = '-'.join(line_list) # join each word in the line_list which are seperated by a '-'
    return new_line

if __name__ == '__main__':
    line = input() # getting the input from the user
    result = split_and_join(line) # calling the function
    print(result) # printing the output
```

## 4. Mutations



```python
def mutate_string(string, position, character):
    return string[:position] + character + string[position+1:] # using the slicing method to get the output

if __name__ == '__main__':
    s = input() # getting the string input
    i, c = input().split() # gettin the position and character to be replaced in the initial string 
    s_new = mutate_string(s, int(i), c) # calling the function
    print(s_new)
```

## 5. Find a string



```python
def count_substring(string, sub_string):
    count = 0
    for idx in range(len(string) + 1 - len(sub_string)): # iterating through the string
        if string[idx:idx+len(sub_string)] == sub_string: # checking whether portion is equal to the sub string
            count += 1 
        else:
            continue 
    return count
        

if __name__ == '__main__':
    string = input().strip() # getting the string input
    sub_string = input().strip() # getting the sub string input to be found in the initial string
    
    count = count_substring(string, sub_string) # calling the function
    print(count)
```

## 7. Text Alignment



```python
thickness = int(input()) #This must be an odd number
c = 'H'

#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

## 8. Text Wrap



```python
import textwrap

def wrap(string, max_width):
    # iterating through the string with a step size of max_width 
    # and returning each string in a new line
    return '\n'.join([string[idx:idx+max_width] for idx in range(0, len(string), max_width)])

if __name__ == '__main__':
    string, max_width = input(), int(input()) # getting the string and width input
    result = wrap(string, max_width) # calling the function
    print(result)
```

## 10. String Formatting



```python
def print_formatted(number):
    l1 = len(bin(number)[2:])
    for num in range(1,number+1): # iterating number + 1 times 
        print(str(num).rjust(l1,' '),end=" ") # printing the string format of number 
        print(oct(num)[2:].rjust(l1,' '),end=" ") # printing the octal format of number
        print(((hex(num)[2:]).upper()).rjust(l1,' '),end=" ") # printing the hexadecimal format of number
        print(bin(num)[2:].rjust(l1,' '),end=" ") # printing the binary format of number
        print("")

if __name__ == '__main__':
    n = int(input()) # getting the number input
    print_formatted(n)
```

## 11. Alphabet Rangoli



```python
import string
def print_rangoli(size):
    width = 4 * size - 3
    alpha = string.ascii_lowercase
    for i in list(range(size))[::-1] + list(range(1, size)):
        print('-'.join(alpha[size-1:i:-1] + alpha[i:size]).center(width, '-'))
        
if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

## 14. Merge the Tools!



```python
def merge_the_tools(string, k):
    for idx in range(0, len(string), k): # iterating through the string at a step size of k
        s = '' # empty string
        for char in string[idx : idx+k]: # for each character in the sliced string
            if (char not in s): # checking whether it is present in s
                s += char # appending the char to s
        print(s)
                

if __name__ == '__main__':
    string, k = input(), int(input()) # getting the string and string seperator input
    merge_the_tools(string, k) # calling the function
```
