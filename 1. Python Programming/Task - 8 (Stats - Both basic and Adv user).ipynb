## 1. Binomial Distribution I



```python
def factorial(n): # function to find the factorial
    return 1 if n == 0 else n*factorial(n-1)

def combination(n, x): # function to find the combination value
    return factorial(n) / (factorial(x) * factorial(n-x))

def binomial_distribution(x, n, p): # function to find the binomial distribution
    return combination(n, x) * p**x * (1-p)**(n-x)

l, r = list(map(float, input().split())) # getting the inputs
odds = l / r
print(round(sum([binomial_distribution(i, 6, odds / (1 + odds)) for i in range(3, 7)]), 3)) # calling the function in a loop
```

## 3. Normal Distribution I



```python
import math

def normal_distribution(x, mean_, std_): # function to find normal distribution
    return round(0.5 * (1 + math.erf((x - mean_) / (std_ * (2 ** 0.5)))), 3)

mean, std_dev = map(float, input().split()) # getting the mean and standard deviation
input_1 = float(input()) # getting the x value
input_2, input_3 = map(float, input().split()) # getting the range value

print(normal_distribution(input_1, mean, std_dev))
print(normal_distribution(input_3, mean, std_dev) - normal_distribution(input_2, mean, std_dev))
```

## 5. The Central Limit Theorem I



```python
import math

def central_limit(mean_, std_, maximum, n): # function to find the central limit
    new_mean = n * mean_ # finding the new mean
    new_std = math.sqrt(n) * std_ # finding the new standard deviation
    return round(0.5 * (1 + math.erf((maximum - new_mean)/(new_std * (2**0.5)))), 4) # calculating the value


max_val = int(input()) # maximum value 
load_val = int(input()) # given value
mean = int(input()) # mean 
std_dev = int(input()) # standard deviation

print(central_limit(mean, std_dev, max_val, load_val))
```

## 8. Pearson Correlation Coefficient I



```python
N = int(input()) # getting the size of the data
X = list(map(float,input().strip().split())) # input list 1
Y = list(map(float,input().strip().split())) # input list 2

mu_x = sum(X) / N # mean of x data
mu_y = sum(Y) / N # mean of y data

stdv_x = (sum([(i - mu_x)**2 for i in X]) / N)**0.5 # standard deviation of x data
stdv_y = (sum([(i - mu_y)**2 for i in Y]) / N)**0.5 # standard deviation of y data


covariance = sum([(X[i] - mu_x) * (Y[i] -mu_y) for i in range(N)]) # calculating the covariance

correlation_coefficient = covariance / (N * stdv_x * stdv_y) # calcultaing the correlation

print(round(correlation_coefficient,3))
```
