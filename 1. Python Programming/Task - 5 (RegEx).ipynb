## 10. HTML Parser - Part 1



```python
import re

tag_filter = re.compile(r"(?<=<)(?:.*?)(?=>)") # expression to evaluate tags
attrib_filter = re.compile(r"([\w-]+)(?:=[\"'](.*?)[\"'])?") # expression to evaluate attribute

s = ""

# get input stream
for _ in range(int(input())):
    s += input()

# discard comments
s = re.sub(r"(?:<!--(.|\s)*?(?:-->))","",s)

# extract individual tag information
tags = tag_filter.findall(s)

# iterate over the extracted tag information
for tag in tags:
    # tag handling logic
    if tag[0] == "/":
        print("End   :",tag[1:])
    else:
        attribs = attrib_filter.findall(tag)
        if tag[-1] == "/":
            print("Empty :",attribs[0][0])
        else:
            print("Start :",attribs[0][0])
            
        for i in range(1,len(attribs)):
            a = attribs[i][0]
            if attribs[i][1] == '': 
                v = 'None' 
            else: 
                v = attribs[i][1]
            print("-> {} > {}".format(a,v))
```

## 11. HTML Parser - Part 2



```python
from html.parser import HTMLParser

class HT(HTMLParser):
    def handle_comment(self, data): # function to handle the comments
        if '\n' in data: 
            print ('>>> Multi-line Comment')
        else:
            print('>>> Single-line Comment')
        print (data)
        
    def handle_data(self, data): # function to evaluate data
        if data !='\n':
            print('>>> Data')
            print(data)
            
html = ""       
for i in range(int(input())):
    html += input().rstrip()
    html += '\n'
    
parser = HT()
parser.feed(html)
parser.close()
```

## 12. Detect HTML Tags, Attributes and Attribute Values



```python
import re

text = ''
for _ in range(int(input())):
    text = re.sub(r'<!.+-->',r' ',(text+input())) # finding the expression to be substituted

for er in re.findall(r'<([^/][^>]*)>', text): # for every character in the expression
    if ' ' in er: # checking if blank space is in er     
        for ere in re.findall(r'([a-z]+)? *([a-z-]+)="([^"]+)', er): # for every character in the expression
            if ere[0]:
                print(ere[0])          
            print('-> '+ere[1]+' > '+ere[2])
    else:
        print(er)
```

## 13. Validating UID



```python
import re

for i in range(int(input())): # iterating n times
    N = input().strip() # getting the input
    if N.isalnum() and len(N) == 10: # checking whether N is alphanumneric or N is of length 10
        if bool(re.search(r'(.*[A-Z]){2,}',N)) and bool(re.search(r'(.*[0-9]){3,}',N)): # searching for the given expression in the input N
            if re.search(r'.*(.).*\1+.*',N): # searching for the expression in the input
                print('Invalid')
            else:
                print('Valid')
        else:
            print('Invalid')
    else:
        print('Invalid')
```

## 14. Validating Credit Card Numbers



```python
import re
for i in range(int(input())): # iterating n times
    S = input().strip() # getting the input
    pre_match = re.search(r'^[456]\d{3}(-?)\d{4}\1\d{4}\1\d{4}$',S) # searching for the expression in the input
    if pre_match: # if matches
        processed_string = "".join(pre_match.group(0).split('-')) # joining the group together 
        final_match = re.search(r'(\d)\1{3,}',processed_string) # searching for the expression in the processed string
        if final_match:
            print('Invalid')
        else :
            print('Valid')
    else:
        print('Invalid')
```

## 16. Matrix Script



```python
first_multiple_input = input().rstrip().split() # getting the input

n = int(first_multiple_input[0]) # getting the input

m = int(first_multiple_input[1]) # getting the input

matrix = [] # empty list

for _ in range(n): # iterating n times
    matrix_item = input() # getting the input
    matrix.append(matrix_item) # appending the input to list

encoded_string = "".join([matrix[j][i] for i in range(m) for j in range(n)]) # joining the list as a string
pat = r'(?<=[a-zA-Z0-9])[^a-zA-Z0-9]+(?=[a-zA-Z0-9])' # expression to be evaluated
print(re.sub(pat,' ',encoded_string)) # substituting the patter in the string
```

## 17. Detect Floating Point Number



```python
import re

n = int(input())
for _ in range(n):
    input_ = input()
    print(bool(re.match(r'^[+-]?[0-9]*\.[0-9]+$', input_)))
    # ^[+-]? ----> starting with + or - that can appear 0 or 1 time
    # [0-9]* ----> set of numbers from 0 to 9 that can appear 0 or more times
    # \. -----> mentions the decimal point
    # [0-9]+$ ----> string that is ending with 0 - 9 that can occour more than time
```
