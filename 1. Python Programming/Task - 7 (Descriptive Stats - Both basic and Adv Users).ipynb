## Mean:

**Mean** can be defined as the ratio of sum of all the values it the dataset and the total number of observations for that particular dataset. It can be given as

$$ \Large \sum_{i=1}^{N} = \frac{X_i} {n}$$

Where   
* $ X_i $ is the individual data points.  
* *n* is the total number of data points

It can also be written as

$$ \Large Mean(X) = \frac{Sum \ of\ observations}{Total \ number \ of \ observations} $$

### Without using library functions:

```python
# here to find the sum of all the data points we can use the inbuilt method sum()
# to find the total number of data points we can use the inbuilt method len()
# we can make use of the column 'Mthly_HH_Income' present in the dataset.

monthly_income_mean = sum(data['Mthly_HH_Income'])/len(data['Mthly_HH_Income'])
monthly_income_mean
```

## Median:

**Median** can be defined as the middle element in a sorted set of data.


$$ \Large
Med(X) = \left\{
    \begin{array}\\
        X[\frac{n}{2}] & \mbox{if } \ n \ is \ Odd  \\
        \frac{X[\frac{n}{2}] + X[\frac{n-1}{2}]}{2} & \mbox{if } \ n \ is \ Even \\
    \end{array}
\right.
$$

Where   
* $ X $ represents the ordered dataset.  
* $ n $ is the total number of data points.

### Without using library functions:

```python
# without using function

monthly_exp = list(data['Mthly_HH_Expense']) # storing the elements in the form of list
monthly_exp.sort() # sorting the list
n = len(monthly_exp) # findfing the total number of elements in the list

if len(monthly_exp) % 2 != 0: # checking if there are odd number of elements in the list
    monthly_exp_median = monthly_exp[n//2] # finding the middle element
else: # if there are even number of elements in the list
    # finding the two middle elements
    m1 = monthly_exp[n//2] 
    m2 = monthly_exp[n//2 - 1]
    # finding the average of the two medians
    monthly_exp_median = (m1+m2)/2
    
print(monthly_exp_median)
```

### Using library functions:

```python
# using library functions
# we can use the np.median() to find the median for a set of data

monthly_exp_func = np.median(data['Mthly_HH_Expense'])
monthly_exp_func
```

## Variance:

**Variance** can be defined as the differences between each number in the data set and the mean, then squaring the differences to make them positive, and finally dividing the sum of the squares by the number of values in the data set.

$$
\Large \sigma^2 = \sum_{i=1}^{N} \frac{(x_i -\mu)^2}{N}
$$

Where 
* $ \sigma^2 $ is the variance of the data
* $ x_i $ is the value of the ith element
* $ \mu $ is the mean of the data
* $ N $ is the size of the data

### Without using library functions:

```python
# without using library functions

# we can use the 'Emi_or_Rent_Amt' column to find the variance 

Emi_or_Rent_Amt_ls = list(data['Emi_or_Rent_Amt']) # converting the series data to list
Emi_or_Rent_Amt_mean = sum(Emi_or_Rent_Amt_ls)/len(Emi_or_Rent_Amt_ls) # finding the average of the list
# finding the variance manually 
Emi_or_Rent_Amt_var = sum((value - Emi_or_Rent_Amt_mean)**2 for value in Emi_or_Rent_Amt_ls)/len(Emi_or_Rent_Amt_ls)
print(Emi_or_Rent_Amt_var)
```

## Standard Deviation:

**Standard Deviation** is a measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean of the set, while a high standard deviation indicates that the values are spread out over a wider range.

$$
\Large \sigma = \sqrt{\frac{\sum(x_i -\mu)^2}{N}}
$$

Where 
* $ \sigma $ is the standard deviation of the data
* $ x_i $ is the value of the ith element
* $ \mu $ is the mean of the data
* $ N $ is the size of the data

### Without using library functions:

```python
# without using functions

# we can use the 'Emi_or_Rent_Amt' column to find the variance

Emi_or_Rent_Amt_ls = list(data['Emi_or_Rent_Amt']) # converting the series data to list
Emi_or_Rent_Amt_mean = sum(Emi_or_Rent_Amt_ls)/len(Emi_or_Rent_Amt_ls) # finding the average of the list
# finding the variance manually 
Emi_or_Rent_Amt_var = sum((value - Emi_or_Rent_Amt_mean)**2 for value in Emi_or_Rent_Amt_ls)/len(Emi_or_Rent_Amt_ls)
Emi_or_Rent_Amt_std = Emi_or_Rent_Amt_var ** 0.5 # finding the square root of variance
print(Emi_or_Rent_Amt_std)
```

### Using library functions:

## Correlation:

* **Correlation** is a statistic that measures the degree to which two variables move in relation to each other.  
* **Correlation** shows the strength of a relationship between two variables and is expressed numerically by the **correlation coefficient**. The correlation coefficient's values range between **-1.0 and 1.0**.
* It can be represented as


$$
\Large r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum(x_i - \bar{x})^2 \sum(y_i - \bar{y})^2}}
$$

Where 
* $ r $ it the correlation coefficient
* $ x_i $ values of the x-variable in the dataset
* $ \bar{x} $ mean of the values of the x-variable
* $ y_i $ values of the y-variable in the dataset
* $ \bar{y} $ mean of the values of the y-variable
$

### Without using library functions:

```python
# without using functions
# we can find the correlation between the columns 'Mthly_HH_Income' and 'Mthly_HH_Expense'

# storing the data points as a list
X = list(data['Mthly_HH_Income'])
Y = list(data['Mthly_HH_Expense'])

# finding the mean of the X and Y
X_mean = sum(X)/len(X)
Y_mean = sum(Y)/len(Y)

# finding the standard deviation
X_std = (sum([(item - X_mean)**2 for item in X]) / len(X)) ** 0.5
Y_std = (sum([(item - Y_mean)**2 for item in Y]) / len(Y)) ** 0.5

# finding the covariance of the data
covariance = sum([(X[i] - X_mean) * (Y[i] - Y_mean) for i in range(len(X))])

# calcultaing the correlation
correlation_coefficient = covariance / (len(X) * X_std * Y_std)

print(round(correlation_coefficient,3))
```

## Normal Distribution:

**Normal Distribution** a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean.


It can be represented as
$$
\Large f(x,\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}} e^{\frac{-1}{2}(\frac{x - \mu}{\sigma})^2}
$$

Where 
* $ x $ is the variable
* $ \mu $ is the mean
* $ \sigma $ is the standard deviation

```python
# creating a normal distribution from random points

from scipy.stats import norm
import statistics

# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(-20, 20, 0.01)
  
# Calculating mean and standard deviation
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)
  
plt.plot(x_axis, norm.pdf(x_axis, mean, sd)) # plotting the distribution
plt.title('Normal Distribution')
plt.show();
```

## Features of Normal Distribution:

* Bell shaped, unimodal and symmetrical in nature.
* The mean, median and mode are exactly the same.
* The distribution is symmetric about the meanhalf the values fall below the mean and half above the mean.
* The distribution can be described by two values: the mean and the standard deviation.



## Skewness:

Skewness is a measure of the symmetry of a distribution. The highest point of a distribution is its mode. The mode marks the response value on the x-axis that occurs with the highest probability. A distribution is skewed if the tail on one side of the mode is fatter or longer than on the other: it is asymmetrical.

It can be represented as

$$
\Large \tilde\mu_3 = \frac{\sum_{i=1}^{N}(X_i - \bar{X})^3}{(N-1)\sigma^3}
$$

Where
* $\tilde\mu_3$ is the  skewness
* $ N $ is the  number of variables in the distribution
* $ X_i $ is the  random variable
* $ \bar{X} $ is the  mean of the distribution
* $ \sigma$ is the standard deviation


## Positive Skewness:

Positive Skewness means when the tail on the right side of the distribution is longer or fatter. The mean and median will be greater than the mode. 

## Negative Skewness:

Negative Skewness is when the tail of the left side of the distribution is longer or fatter than the tail on the right side. The mean and median will be less than the mode.

##  Box Cox Plot:

A Box cox transformation is defined as a way to transform non-normal dependent variables in our data to a normal shape through which we can run a lot more tests than we could have.

```python
# box cox plot for random points

# generate non-normal data (exponential)
original_data = np.random.exponential(size = 1000)

# transform training data & save lambda value
fitted_data, fitted_lambda = stats.boxcox(original_data)

# creating axes to draw plots
fig, ax = plt.subplots(1, 2)

# plotting the original data(non-normal) and
# fitted data (normal)
sns.distplot(original_data, hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2},
            label = "Non-Normal", color ="green", ax = ax[0])

sns.distplot(fitted_data, hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2},
            label = "Normal", color ="green", ax = ax[1])

# adding legends to the subplots
plt.legend(loc = "upper right")

# rescaling the subplots
fig.set_figheight(5)
fig.set_figwidth(10)

print(f"Lambda value used for Transformation: {fitted_lambda}")
plt.tight_layout()
```

```python
for col in data.columns[:2]:
    if data[col].dtype != 'object':
        # generate non-normal data (exponential)
        original_data = np.array(data[col])
        # transform training data & save lambda value
        fitted_data, fitted_lambda = stats.boxcox(original_data)

        # creating axes to draw plots
        fig, ax = plt.subplots(1, 2)

        # plotting the original data(non-normal) and
        # fitted data (normal)
        sns.distplot(original_data, hist = False, kde = True,
                kde_kws = {'shade': True, 'linewidth': 2},
                label = "Non-Normal", color ="green", ax = ax[0])

        sns.distplot(fitted_data, hist = False, kde = True,
                kde_kws = {'shade': True, 'linewidth': 2},
                label = "Normal", color ="green", ax = ax[1])
        plt.title('{}'.format(col))

        # adding legends to the subplots
        plt.legend(loc = "upper right")

        # rescaling the subplots
        fig.set_figheight(5)
        fig.set_figwidth(10)

        print(f"Lambda value used for Transformation: {fitted_lambda}")
        plt.tight_layout()
```
