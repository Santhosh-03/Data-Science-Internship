## 1. Arrays



```python
import numpy # importing the numpy library
def arrays(arr):
    arr = arr[::-1] # reversing the list
    rev_arr = numpy.array(arr, float) # converting the list to a numpy array
    return rev_arr
    
arr = input().strip().split(' ') # getting the list input
result = arrays(arr) # calling the function and passing the list as argument
print(result)
```

## 2. Shape and Reshape



```python
import numpy as np # imporing the numpy library

arr = list(map(int, input().split())) # getting the list input
numpy_arr = np.array(arr).reshape(3,3) # converting the list as a numpy array and changing its shape to 3X3
print(numpy_arr)
```

## 3. Transpose and Flatten



```python
import numpy as np # importing the numpy library

row, col = map(int, input().split()) # getting the dimension of the array

numpy_arr = np.array([input().strip().split() for _ in range(row)], int) # getting the array input
print(numpy_arr.T) # transpose of the array
print(numpy_arr.flatten()) # flattened 1-D array
```

## 4. Concatenate



```python
import numpy as np # importing the library

# n - size of 1st array
# m - size of 2nd array
# p - number of elements in the row
n, m, p = map(int,input().split()) 
a1 = np.array([input().split() for _ in range(n)], int) # getting the first array input
a2 = np.array([input().split() for _ in range(m)], int) # getting the first array input
print(np.concatenate((a1, a2), axis = 0))  # concatenating the two arrays one below the other
```

## 7. Array Mathematics



```python
import numpy as np # importing the library

row, col = map(int, input().split()) # getting the number of rows and columns
arr1 = np.array([input().strip().split() for _ in range(row)], int) # getting the first array
arr2 = np.array([input().strip().split() for _ in range(row)], int) # getting the second array

# printing the mathematical operations
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 // arr2)
print(arr1 % arr2)
print(arr1 ** arr2)
```

## 8. Floor, Ceil and Rint



```python
import numpy as np # importing the library
np.set_printoptions(legacy = '1.13') # setting the printing options

arr = list(map(float, input().split())) # getting the list input

numpy_arr = np.array(arr, float) # converting the list to a numpy array
print(np.floor(numpy_arr)) # printing the floor values
print(np.ceil(numpy_arr)) # printing the ceil values
print(np.rint(numpy_arr)) # printing the rounded values
```

## 9. Sum and Prod



```python
import numpy as np # importing the library

row, col = map(int, input().split()) # getting the number of rows and columns
numpy_arr = np.array([input().strip().split() for _ in range(row)], int) # getting the array

row_sum = np.sum(numpy_arr, axis = 0) # finding the sum of elements along the columns
prod = np.prod(row_sum) # finding the product of elements in the sum array
print(prod)
```

## 10. Min and Max



```python
import numpy as np # importing the library

row, col = map(int, input().split()) # getting the number of rows and columns
numpy_arr = np.array([input().strip().split() for _ in range(row)], int) # getting the input array
min_arr = np.min(numpy_arr, axis = 1) # finding the minimum element along the rows of the array
max_num = np.max(min_arr, axis = 0) # finding the maximum element from the minimum elements of the array
print(max_num)
```

## 11. Mean, Var, and Std



```python
import numpy as np # importing the library

row, col = map(int, input().split()) # getting the number of rows and columns
numpy_arr = np.array([input().strip().split() for _ in range(row)], int) # getting the numpy array

print(np.mean(numpy_arr, axis = 1)) # finding the mean along the columns
print(np.var(numpy_arr, axis = 0)) # finding the variance along the rows
print(round(np.std(numpy_arr, axis = None),11)) # finding the standard variation for all the elements
```

## 12. Dot and Cross

## 13. Inner and Outer



```python
import numpy as np # importing the libraries

a = list(map(int, input().split())) # getting the list input
b = list(map(int, input().split())) # getting the list input
print(np.inner(np.array(a), np.array(b))) # inner product of two arrays
print(np.outer(np.array(a), np.array(b))) # outer product of two arrays
```
