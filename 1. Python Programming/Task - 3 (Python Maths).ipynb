## 1. Polar Coordinates



```python
import cmath

num = complex(input())
print(abs(complex(num.real, num.imag))) # z value which is the distance
print(cmath.phase(complex(num.real, num.imag))) # counter clock wise angle value
```

## 2. Find Angle MBC



```python
import math
 
ab = int(input()) # getting the value of side 1
bc = int(input()) # getting the value of side 2
ca = math.hypot(ab, bc) # getting the hypotenuse using the 2 sides
angle = round(math.degrees(math.acos(bc/ca))) # calculating the angle mbc in the triangle
deg = chr(176) # used to print degree symbol
print(angle, deg, sep = '')
```

## 3. Triangle Quest 2



```python
n = int(input()) # getting the number of rows
for i in range(1, n+1):
    print(int(((10**i-1)/9)**2)) # calculating the square values of 1,11,111,1111...
```

## 4. Mod Divmod



```python
num1 = int(input()) # getting the dividend
num2 = int(input()) # getting the divisor
quotient, remainder = divmod(num1,num2) # getting the quotient and remainder
print(quotient)
print(remainder)
print(tuple((quotient, remainder)))
```

## 5. Power - Mod Power



```python
a = int(input())
b = int(input())
m = int(input())

print(pow(a,b)) # calulating the value of a ** b
print(pow(a,b,m)) # calulating the value of ((a ** b)%m) and finding the remainder
```
